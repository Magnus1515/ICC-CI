```python
class Treenode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.middle = None  # New middle node
        self.right = None

class Tree:
    def __init__(self):
        self.root = None

def height(root):
    if root is None:
        return 0
    # Include height calculation for the middle node
    return max(height(root.left), height(root.middle), height(root.right)) + 1

def getcol(h):
    # The number of columns in the bottom row of a full ternary tree is
    # 3 nodes for the first row, and each subsequent row has the previous
    # row's count multiplied by 3 and then subtract 2 for spacing
    if h == 1:
        return 1
    return 3 * getcol(h - 1) - 2


def printTree(M, root, col, row, height):
    if root is None or col < 0 or row < 0:
        return
    M[row][col] = root.data
    gap = max(0, pow(3, height - 2) - 1)  # Ensure there is no negative gap
    printTree(M, root.left, col - gap, row + 1, height - 1)
    printTree(M, root.middle, col, row + 1, height - 1)
    printTree(M, root.right, col + gap, row + 1, height - 1)


def TreePrinter():
    h = height(myTree.root)
    col = getcol(h)
    M = [[0 for _ in range(col)] for __ in range(h)]
    printTree(M, myTree.root, col//2, 0, h)
    for i in M:
        for j in i:
            if j == 0:
                print(" ", end=" ")
            else:
                print(j, end=" ")
        print("")

# Sample usage
myTree2 = Tree()
myTree2.root = Treenode(1)
myTree2.root.left = Treenode(2)
myTree2.root.middle = Treenode(3)  # Adding middle nodes
myTree2.root.right = Treenode(4)

TreePrinter()
```
