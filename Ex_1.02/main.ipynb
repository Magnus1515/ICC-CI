```python
import matplotlib.pyplot as plt 
import csv

maze = []
with open("maze_01.csv", 'r') as file:
    for line in file:
        line = line.rstrip()
        row = []
        for c in line:
            if c == ' ':
                row.append(0) # spaces are 1s
            else:
                row.append(1) # walls are 0s
        maze.append(row)

 

with open('maze_01.csv', mode ='r')as file:
	
    csvFile = csv.reader(file)
    for lines in csvFile:
		    print(lines)


```

```python
import matplotlib.pyplot as plt
import csv

maze = []
with open("maze_01.csv", 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        maze.append([int(cell) for cell in row])

plt.imshow(maze, cmap='binary')
plt.xticks([]) 
plt.yticks([]) 
plt.show()
```
